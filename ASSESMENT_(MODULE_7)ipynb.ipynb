```python
# 1) HEALTH CARE iNDUSTRY #
class ClinicAppointment:
    def __init__(self):
        # Dictionary to store appointments: {doctor: {time_slot: [patients]}}
        self.appointments = {}
        self.time_slots = ["10am", "11am", "12pm", "2pm", "3pm"]
        self.max_per_slot = 3

    def book_appointment(self):
        print("\n--- Book Appointment ---")
        name = input("Enter patient name: ")
        age = input("Enter patient age: ")
        mobile = input("Enter mobile number: ")
        doctor = input("Enter preferred doctor: ")

        # Initialize doctor if not present
        if doctor not in self.appointments:
            self.appointments[doctor] = {slot: [] for slot in self.time_slots}

        # Show available slots
        print("Available time slots:")
        for slot in self.time_slots:
            booked = len(self.appointments[doctor][slot])
            print(f"{slot} - {self.max_per_slot - booked} available")

        slot_choice = input("Choose your time slot: ")
        if slot_choice not in self.time_slots:
            print("Invalid time slot.")
            return

        if len(self.appointments[doctor][slot_choice]) >= self.max_per_slot:
            print("Sorry, this time slot is full. Please choose another.")
            return

        # Book the appointment
        patient = {"name": name, "age": age, "mobile": mobile}
        self.appointments[doctor][slot_choice].append(patient)
        print(f"Appointment confirmed for {name} with Dr.{doctor} at {slot_choice}.")

    def view_appointment(self):
        print("\n--- View Appointment ---")
        mobile = input("Enter your mobile number: ")
        found = False
        for doctor, slots in self.appointments.items():
            for time, patients in slots.items():
                for patient in patients:
                    if patient["mobile"] == mobile:
                        print(f"Appointment: Dr.{doctor}, {time}, Patient: {patient['name']}, Age: {patient['age']}")
                        found = True
        if not found:
            print("No appointment found for this mobile number.")

    def cancel_appointment(self):
        print("\n--- Cancel Appointment ---")
        mobile = input("Enter your mobile number: ")
        for doctor, slots in self.appointments.items():
            for time, patients in slots.items():
                for i, patient in enumerate(patients):
                    if patient["mobile"] == mobile:
                        del patients[i]
                        print(f"Appointment for {patient['name']} with Dr.{doctor} at {time} has been canceled.")
                        return
        print("No appointment found to cancel.")

    def run(self):
        while True:
            print("\n--- Clinic Appointment System ---")
            print("1. Book Appointment")
            print("2. View Appointment")
            print("3. Cancel Appointment")
            print("4. Exit")
            choice = input("Enter your choice: ")

            if choice == "1":
                self.book_appointment()
            elif choice == "2":
                self.view_appointment()
            elif choice == "3":
                self.cancel_appointment()
            elif choice == "4":
                print("Exiting system. Goodbye!")
                break
            else:
                print("Invalid choice. Try again.")

# Run the system
clinic = ClinicAppointment()
clinic.run()
```

```python
### School Managment System ###
class SchoolManagement:
    def __init__(self):
        self.students = {}  # Store student records as {student_id: info}
        self.next_id = 1    # Auto-generate unique student IDs


    #  New Admission

    def new_admission(self):
        print("\n--- New Admission ---")
        name = input("Enter Student Name: ")
        try:
            age = int(input("Enter Age (5–18): "))
        except ValueError:
            print(" Invalid age input. Please enter a number.")
            return

        if not (5 <= age <= 18):
            print(" Age must be between 5 and 18.")
            return

        try:
            student_class = int(input("Enter Class (1–12): "))
        except ValueError:
            print(" Invalid class input. Please enter a number.")
            return

        if not (1 <= student_class <= 12):
            print(" Class must be between 1 and 12.")
            return

        mobile = input("Enter Guardian's Mobile Number (10 digits): ")
        if not (mobile.isdigit() and len(mobile) == 10):
            print(" Invalid mobile number. Must be 10 digits.")
            return

        # Assign student ID
        student_id = self.next_id
        self.students[student_id] = {
            "Name": name,
            "Age": age,
            "Class": student_class,
            "Mobile": mobile
        }
        self.next_id += 1

        print(f" Admission Successful! Assigned Student ID: {student_id}")


    #  View Student Details

    def view_student(self):
        print("\n--- View Student Details ---")
        try:
            student_id = int(input("Enter Student ID: "))
        except ValueError:
            print(" Invalid input. Please enter a number.")
            return

        student = self.students.get(student_id)
        if student:
            print("\n--- Student Information ---")
            for key, value in student.items():
                print(f"{key}: {value}")
        else:
            print(" Student ID not found.")


    #  Update Student Info

    def update_student(self):
        print("\n--- Update Student Info ---")
        try:
            student_id = int(input("Enter Student ID to update: "))
        except ValueError:
            print(" Invalid input. Please enter a number.")
            return

        if student_id not in self.students:
            print(" Student ID not found.")
            return

        print("1. Update Class")
        print("2. Update Mobile Number")
        choice = input("Enter choice (1 or 2): ")

        if choice == '1':
            try:
                new_class = int(input("Enter New Class (1–12): "))
            except ValueError:
                print(" Invalid class input.")
                return

            if not (1 <= new_class <= 12):
                print(" Class must be between 1 and 12.")
                return
            self.students[student_id]['Class'] = new_class
            print(" Class updated successfully!")

        elif choice == '2':
            new_mobile = input("Enter New Mobile Number (10 digits): ")
            if not (new_mobile.isdigit() and len(new_mobile) == 10):
                print(" Invalid mobile number.")
                return
            self.students[student_id]['Mobile'] = new_mobile
            print(" Mobile number updated successfully!")
        else:
            print(" Invalid choice.")

    #  Remove Student Record

    def remove_student(self):
        print("\n--- Remove Student Record ---")
        try:
            student_id = int(input("Enter Student ID to remove: "))
        except ValueError:
            print(" Invalid input.")
            return

        if student_id in self.students:
            del self.students[student_id]
            print(" Student record removed successfully!")
        else:
            print(" Student ID not found.")

    #  Exit System

    def exit_system(self):
        print("\n👋 Exiting School Management System. Goodbye!")
        exit()



# Main Program

def main():
    system = SchoolManagement()

    while True:
        print("\n====== SCHOOL MANAGEMENT SYSTEM ======")
        print("1. New Admission")
        print("2. View Student Details")
        print("3. Update Student Info")
        print("4. Remove Student Record")
        print("5. Exit")
        choice = input("Enter your choice (1–5): ")

        if choice == '1':
            system.new_admission()
        elif choice == '2':
            system.view_student()
        elif choice == '3':
            system.update_student()
        elif choice == '4':
            system.remove_student()
        elif choice == '5':
            system.exit_system()
        else:
            print(" Invalid choice. Please try again.")


if __name__ == "__main__":
    main()
```

```python
### Transport Reservation System (Bus Ticketing) ###

class BusReservation:
    def __init__(self):
        # Predefined routes with prices
        self.routes = {
            "Mumbai to Pune": 500,
            "Delhi to Jaipur": 600,
            "Bangalore to Mysore": 300,
            "Chennai to Coimbatore": 350,
            "Lucknow to Kanpur": 200
        }

        # Store booked tickets: {ticket_id: ticket_info}
        self.tickets = {}

        # Keep track of booked seats per route: {route: [seat_numbers]}
        self.booked_seats = {route: [] for route in self.routes}

        self.next_ticket_id = 1  # Unique ticket ID counter


    #  Show Available Routes

    def show_routes(self):
        print("\n--- Available Bus Routes ---")
        for route, price in self.routes.items():
            print(f"{route} - ₹{price}")


    #  Book Ticket

    def book_ticket(self):
        print("\n--- Book a Ticket ---")
        name = input("Enter Passenger Name: ")

        try:
            age = int(input("Enter Age: "))
        except ValueError:
            print(" Invalid age. Please enter a number.")
            return

        mobile = input("Enter Mobile Number (10 digits): ")
        if not (mobile.isdigit() and len(mobile) == 10):
            print(" Invalid mobile number.")
            return

        print("\nAvailable Routes:")
        for route in self.routes:
            print(f"- {route}")

        route = input("Enter Route (as shown above): ")
        if route not in self.routes:
            print(" Invalid route.")
            return

        # Check seat availability
        if len(self.booked_seats[route]) >= 40:
            print(" No seats available for this route.")
            return

        # Assign seat number
        seat_no = len(self.booked_seats[route]) + 1
        self.booked_seats[route].append(seat_no)

        # Generate unique ticket ID
        ticket_id = self.next_ticket_id
        self.next_ticket_id += 1

        # Store ticket details
        self.tickets[ticket_id] = {
            "Name": name,
            "Age": age,
            "Mobile": mobile,
            "Route": route,
            "Seat No": seat_no,
            "Price": self.routes[route]
        }

        print(f"\n Ticket Booked Successfully!")
        print(f" Ticket ID: {ticket_id}")
        print(f" Seat Number: {seat_no}")
        print(f" Fare: ₹{self.routes[route]}")

    #  view Ticket

    def view_ticket(self):
        print("\n--- View Ticket ---")
        try:
            ticket_id = int(input("Enter Ticket ID: "))
        except ValueError:
            print(" Invalid Ticket ID.")
            return

        ticket = self.tickets.get(ticket_id)
        if ticket:
            print("\n--- Ticket Details ---")
            for key, value in ticket.items():
                print(f"{key}: {value}")
        else:
            print(" Ticket not found.")

    #  Cancel Ticket

    def cancel_ticket(self):
        print("\n--- Cancel Ticket ---")
        try:
            ticket_id = int(input("Enter Ticket ID to Cancel: "))
        except ValueError:
            print(" Invalid input.")
            return

        if ticket_id in self.tickets:
            route = self.tickets[ticket_id]['Route']
            seat_no = self.tickets[ticket_id]['Seat No']

            # Remove seat from booked list
            self.booked_seats[route].remove(seat_no)

            # Remove ticket
            del self.tickets[ticket_id]
            print(" Ticket Cancelled Successfully!")
        else:
            print(" Ticket ID not found.")


    #  Exit System

    def exit_system(self):
        print("\n Exiting Bus Reservation System. Goodbye!")
        exit()


# --------------------------
# Main Program
# --------------------------
def main():
    system = BusReservation()

    while True:
        print("\n====== BUS RESERVATION SYSTEM ======")
        print("1. Show Available Routes")
        print("2. Book Ticket")
        print("3. View Ticket")
        print("4. Cancel Ticket")
        print("5. Exit")
        choice = input("Enter your choice (1–5): ")

        if choice == '1':
            system.show_routes()
        elif choice == '2':
            system.book_ticket()
        elif choice == '3':
            system.view_ticket()
        elif choice == '4':
            system.cancel_ticket()
        elif choice == '5':
            system.exit_system()
        else:
            print(" Invalid choice. Please try again.")


if __name__ == "__main__":
    main()

```
